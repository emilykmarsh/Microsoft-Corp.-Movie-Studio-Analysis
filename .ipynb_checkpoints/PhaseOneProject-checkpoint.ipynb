# Opus Dataset Cleaning

```python
#dropping null values from the dataset

opusClean = opusDf.dropna(axis=0)
opusClean.info()
```

```python
#confirmed that all zeroes including zeroes as strings have been dropped from the database

count = (opusClean['production_budget'] == 0).sum()
print('Count of zeros in Column Budget : ', count)
```

```python
#Creating a total box office column with the combined total of domestic and international box office revenue

opusClean.loc[:,'total_box_office'] = (
    opusClean['domestic_box_office'] + opusClean['international_box_office']
            )
opusClean.head()
```

```python
#Creating a column for total profit calculated by subtracting production budget from the total box office

opusClean.loc[:,'total_profit'] = (
    opusClean['total_box_office'] - opusClean['production_budget']
            )
opusClean.head()
```

```python
#Adding a column for percent of return by calculating the percentage of return on investment using budget and profit

opusClean.loc[:,'percent_return'] = (
    (opusClean['total_box_office']-opusClean['production_budget'])/(opusClean['total_box_office'])* 100
    )
opusClean.head()
```

```python
#Adding a column for percent of international box office of total box office

opusClean.loc[:,'percent_of_int'] = (
    opusClean['international_box_office'] / opusClean['total_box_office'] * 100
    )

opusClean.head()
```

```python
#Reviewing added columns to dataset

opusClean.info()
```

```python
# Adding a column to assign a value for profitable or not profitable using percent return on intial budget

# create a list of our conditions
conditions = [
    (opusClean['percent_return'] <= 0),
    (opusClean['percent_return'] > 0)
    ]

# create a list of the values we want to assign for each condition
values = [0, 1]

# create a new column and use np.select to assign values to it using our lists as arguments
opusClean['profitable'] = np.select(conditions, values)

# display updated DataFrame
opusClean.head()
```

```python
#Adding a column to assign a value for a movie having more then fifty percent of its box office coming from international and therefore labeled an international success

# create a list of our conditions
conditions = [
    (opusClean['percent_of_int'] <= 50),
    (opusClean['percent_of_int'] > 50)
    ]

# create a list of the values we want to assign for each condition
values = [0, 1]

# create a new column and use np.select to assign values to it using our lists as arguments
opusClean['international_success'] = np.select(conditions, values)

# display updated DataFrame
opusClean.head()
```

```python
#Exploring the amount of movies that were international successes in the database

opusClean['international_success'].value_counts()
```

# Opus Data Analysis

```python
#exploration of distribution of the data in the Opus dataset

opusClean.describe()
```

```python
#create a variable for average total profit per movie for the entire dataset

opusAvProfit = opusClean['total_profit'].mean()
print(opusAvProfit)
```

```python
# Organizing counts, total box office, international and domestic box offices for summary chart

counts = (
    opusClean
    .groupby('production_year')['movie_odid']
    .count()
    .reset_index()
    )

totalboxyear = (
    opusClean
    .groupby('production_year')['total_box_office']
    .sum()
    .reset_index()
    )

totalinter = (
    opusClean
    .groupby('production_year')['international_box_office']
    .sum()
    .reset_index()
    )

totaldom = (
    opusClean
    .groupby('production_year')['domestic_box_office']
    .sum()
    .reset_index()
    )

result = pd.merge(counts, totalboxyear, how="inner", on=['production_year', 'production_year'])
resultDomInt = pd.merge(totalinter, totaldom, how="inner", on=['production_year', 'production_year'])
resultFinal= pd.merge(result, resultDomInt, how="inner", on=['production_year', 'production_year'])
resultFinal
```

```python
# Creating a stacked bar chart to show total box office profits for each year and how much of that profit was domestic and international

fig,ax  = plt.subplots(figsize=(14,7))

ax.bar(x=resultFinal['production_year'], height=resultFinal['international_box_office'], color=(0.2, 0.4, 0.6, 0.6), label = 'International Box Office' )
ax.bar(x=resultFinal['production_year'], height=resultFinal['domestic_box_office'], label = 'Domestic Box Office')
         
ax.set_xlabel('Year')
ax.set_ylabel('Profit in Dollars')
ax.set_title('Total Box Office International and Domestic Per Year')
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))

plt.xticks(ticks= resultFinal['production_year'], labels = resultFinal['production_year'])

fig.legend(loc='upper right', bbox_to_anchor=(.88,.88))

plt.savefig("SummaryChart.jpg");
```

This chart shows all of the opus box office data by year. Based on this visualization, certain years varied greatly in terms of the overall box office profits. This variation seems to follow the total amount of movies produced in that year. It is also interesting to note that international box office profits have increased since 2016 showing that international audiences have become an important part of generating profit on a movie. 

```python
# Creating a scatter plot to show the correlation between budget and profit

fig,ax  = plt.subplots(figsize=(14,7))
sns.regplot(x=opusClean["production_budget"], y=opusClean["total_profit"])

ax.set_xlabel('Producution Budget')
ax.set_ylabel('Total Profit')
ax.set_title('Budget Versus Profit in the Opus Data Set')

ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))
ax.xaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))

plt.show();
```

In this visulization it can be seen that a significant amount of the movies in the Opus Dataset at least broke even on their initial investment. It should be noted that the budget does not include marketing and box office does not include additional avenues of revenue. This visualization also shows a strong positive correlation between budget and total profit.

# Source Material's Effect on Profitability

Source material for movies seemed to be an interesting factor to look at more closely in this datatset to see it's effect on profitability. 

```python
# Grouping the dataset by source material and calculating the average profit per movie to account for some source materials having more movies in the dataset then others

opusTopSource = (opusClean
                .groupby(['source'])
                .mean()
                .reset_index()
                .sort_values(by='total_profit', 
                    ascending=False)
                .head()
                           )
opusTopSource
```

```python
# graphing the average profit per movie for the top five source materials based on profit

fig,ax  = plt.subplots(figsize=(14,7))

ax.bar(x=opusTopSource['source'], height=opusTopSource['total_profit'], color=(0.2, 0.4, 0.6, 0.6))
         
ax.set_xlabel('Source')
ax.set_ylabel('Average Movie Profit in Dollars')
ax.set_title('Top Five Average Movie Profits By Source Material')
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))

plt.xticks(ticks= opusTopSource['source'], labels = opusTopSource['source'])

plt.savefig("TopSourceMaterial.jpg");
```

Ranking movies based on average profitability per movie gave some interesting results. Source material that references an existing franchise whether that be a theme park ride, toy or spin off of another franchise has a positive impact on profit. This makes sense as the movie will already have audience interest based on the franchise the movie is being based on. This also allows a production company to pick a franchise that already has significant interest and will be able to save on marketing.
